```python
df['x_vv']=dfv.vv_x
df['y_vv']=dfv.vv_y
df['z_vv']=dfv.vv_z
```

### We wanna try to transform from r,v to OE.

Let's try the first row (0):

```python
rr=[df.x[0],df.y[0],df.z[0]]
rr = np.array(rr)
rr # in km
```

```python
vv=[df.x_vv[0],df.y_vv[0],df.z_vv[0]]
vv = np.array(vv)
vv # in km/s
```

```python
p,ecc,inc,raan,argp,nu = rv2coe(k=Earth.k*1e-9, r=rr, v=vv)
```

### Works, now we only need to loop through all of the df and do this for all rows.

```python
df_p = []
df_ecc = []
df_inc = []
df_raan = []
df_argp = []
df_nu = []


for i in range(0,500):
    rrr = np.array([df.x[i],df.y[i],df.z[i]])
    vvv = np.array([df.x_vv[i],df.y_vv[i],df.z_vv[i]])
    p,ecc,inc,raan,argp,nu = rv2coe(k=Earth.k*1e-9, r=rrr, v=vvv)
    
    df_p.append(p)
    df_ecc.append(ecc)
    df_inc.append(inc)
    df_raan.append(raan)
    df_argp.append(argp)
    df_nu.append(nu)
```

```python
df_OE = pd.DataFrame(
    {'p': df_p,
     'ecc': df_ecc,
     'inc': df_inc,
     'raan': df_raan,
     'argp': df_argp,
     'nu': df_nu,
    })
```

### Perfect, seems like values are stored properly, with decimal places preserved and we successfully created a 500 len DF of ORBITAL ELEMENTS from the R and V position and velocity vectors.

## Let's make a function to do this for us.

- Input: x,y,z,x_vv,y_vv,z_vv dataframe (df)
- Output: p,ecc,inc,raan,argp,nu dataframe (df_OE)

```python
df
```

```python
# Function

def transform_rv_to_oe(df): # df as input
    df_p = []
    df_ecc = []
    df_inc = []
    df_raan = []
    df_argp = []
    df_nu = []
    
    #Iterate all rows of R,V datapoints
    for i in range(0,500):
        
        rrr = np.array([df.x[i],df.y[i],df.z[i]])
        vvv = np.array([df.x_vv[i],df.y_vv[i],df.z_vv[i]])
        p,ecc,inc,raan,argp,nu = rv2coe(k=Earth.k*1e-9, r=rrr, v=vvv)
    
        df_p.append(p)
        df_ecc.append(ecc)
        df_inc.append(inc)
        df_raan.append(raan)
        df_argp.append(argp)
        df_nu.append(nu)
    
    # Write to DF with OE
    df_OE = pd.DataFrame(
        {'p': df_p,
         'ecc': df_ecc,
         'inc': df_inc,
         'raan': df_raan,
         'argp': df_argp,
         'nu': df_nu,
        })
    
    #print(df_OE.head())
    return df_OE
```

```python
# Call

transform_rv_to_oe(df)
```

```python
# Call assigned

df_OE=transform_rv_to_oe(df)
df_OE
```
